# Inventory & Billing System

```python
import mysql.connector
```

```python
def connect_db():
    conn = mysql.connector.connect(
        host = 'localhost',
        user = 'root',
        password = 'RishabhMahar',
        database = 'Inventory_System'
    )
    return conn
```

# Add New Products


```python
def Add_Product(conn):
    cursor = conn.cursor()  

    Name = input("Enter The Product Name : ")
    Price = float(input("Enter The Price Of The Product : "))
    Quantity = int(input("Enter The Quantity Of The Product : "))

    query = "INSERT INTO Products (Name, Price, Quantity) VALUES (%s, %s, %s)"
    values = (Name, Price, Quantity)

    cursor.execute(query, values)
    conn.commit()

    print("Product Successfully Added!\n")
```

# Update Stock Quantity

```python
def Update_Stock(conn):
    Id = int(input("Enter The Product Id :"))
    Qty = int(input("Enter The New Quantity :"))
    conn.execute("UPDATE products SET quantity=? WHERE id=?", (Qty, Id))
    conn.commit()
    print("Stock updated!\n")
```

# Delete Items

```python
def Delete_Item(conn):
    Id = int(input("Enter The Product Id :"))
    conn.execute("DELETE INTO Products WHERE Id = ?",(Id))
    conn.commit()
    print("Item Deleted!\n")
```

# Generate Customer Bill

```python
def Generate_Bill(conn):
    Id = int(input("Enter Product ID: "))
    Qty = int(input("Enter quantity: "))

    cur = conn.cursor()

    cur.execute("SELECT name, price, quantity FROM Products WHERE id=%s", (Id,))
    product = cur.fetchone()

    if product:
        if product[2] >= Qty:  # stock check
            total = product[1] * Qty
            print(f"\n----- BILL -----")
            print(f"Product: {product[0]}")
            print(f"Price: {product[1]}")
            print(f"Quantity: {Qty}")
            print(f"Total: {total}\n")

            # reduce stock
            cur.execute("UPDATE Products SET quantity = quantity - %s WHERE id=%s", (Qty, Id))
            conn.commit()

            # save sale
            cur.execute("INSERT INTO Sales (product_name, quantity, total, date) VALUES (%s, %s, %s, NOW())",
                        (product[0], Qty, total))
            conn.commit()
            print("Bill generated & stock updated!\n")
        else:
            print("Not enough stock!\n")
    else:
        print("Product not found!\n")
```

# View sales history


```python
def view_sales_history(conn):
    cur = conn.cursor()
    cur.execute("SELECT * FROM Sales")
    records = cur.fetchall()
    print("\n---- Sales History ----")
    for row in records:
        print(row)
    print("-----------------------\n")
```

# SQL database connectivity for data storage



```python
def Menu():
    conn = connect_db()
    while True:
        print('----------------Inventory & Billing System-------------------')
        print('1. Add New Product')
        print('2. Update Stock Quantity')
        print('3. Delete Items')
        print('4. Generate Customer Bill')
        print('5. View Sales History')
        print('6. Exit')
        choice = int(input('Enter Your Choice :'))
        if choice == 1:
            Add_Product(conn)
        elif choice == 2:
            Update_Stock(conn)
        elif choice == 3:
            Delete_Item(conn)
        elif choice == 4:
            Generate_Bill(conn)
        elif choice == 5:
            view_sales_history(conn)
        elif choice == 6:
            print("Exiting system...")
            break
        else:
            print("Invalid choice! Try again.\n")
if __name__ == "__main__":
  Menu()
        
```
